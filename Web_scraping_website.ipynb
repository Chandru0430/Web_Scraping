```python
import requests

from bs4 import BeautifulSoup

r = requests.get("https://www.amazon.in/s?i=specialty-aps&srs=205023658031&s=popularity-rank&fs=true&_encoding=UTF8&pf_rd_p=304b0e1e-19e9-4245-92dc-6144eb0ae541&pf_rd_r=NW46B87AJQKAXX0STNY9&ref=cct_cg_Explore_2a1")
c = r.content
print(c)
```

```python
soup=BeautifulSoup(c,"html.parser")
print(soup)
```

```python
all=soup.find_all("div", {"data-component-type": "s-search-result"})

len(all)
```

```python
print(all)
```

```python
h2=all[0].find("h2", {"class": "a-size-medium a-spacing-none a-color-base a-text-normal"})
print(h2)
```

```python
for item in all:   #to display all price value in the website
    print(item.find_all("h2", {"class": "a-size-medium a-spacing-none a-color-base a-text-normal"})[0].text)
    print(item.find_all("div",{"class":"a-row"})[0].text)
    print(item.find_all("span",{"class":"a-offscreen"})[0].text)
```

```python
l=[]
for i in all:   #to display all price value in the website
    d={}
    d["Fridge_name"]=i.find_all("h2", {"class": "a-size-medium a-spacing-none a-color-base a-text-normal"})[0].text
    d["Row_info"]=i.find_all("div",{"class":"a-row"})[0].text
    d["Price"]=i.find_all("span",{"class":"a-offscreen"})[0].text
    l.append(d)
print(l)
```

```python
import pandas
df=pandas.DataFrame(l)
print(df)
df.to_csv("amazon_refridgerator.csv")
```
