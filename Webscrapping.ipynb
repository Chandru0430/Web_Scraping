```python
import requests
url = "https://ticketnew.com/movies/chennai"
response = requests.get(url)
response.content
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
movie_names = soup.find_all('div', class_="DesktopRunningMovie_movTitle__Q1pOY")
languages = soup.find_all('div', class_="DesktopRunningMovie_point__YtD_w")
```

```python
l = []
for i,j in zip(movie_names,languages):
    l.append({'Movie':i.get_text(), 'Rating':j.find('span').get_text(), 'Languages':j.find('span',class_="textClamp").get_text()})
    #print(f"Movie : {i.get_text()}")
    #print(f"Rating : {j.find('span').get_text()}")
    #print(f"Languages : {j.find('span',class_="textClamp").get_text()}")
    #print("="*80)
print(l)
```

```python

import pandas as pd

df = pd.DataFrame(l)
df
```

```python
df.to_csv("Movie_names.csv",index=False)
```

```python
df.to_excel("Movie_names.xlsx", index=False)
```
